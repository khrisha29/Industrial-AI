## Exploratory Data Analysis (EDA) Summary

The dataset consists of time-series sensor readings from industrial robotic equipment, including temperature, vibration, pressure, and engineered temporal features such as lag values and rolling window statistics. The data was verified to be free from missing values after interpolation and preprocessing.

Analysis of the target variable shows a highly imbalanced class distribution, with failure events representing a very small fraction of the total observations. This reflects real-world predictive maintenance scenarios and highlights the need for recall- and F1-focused evaluation metrics in subsequent modeling stages.

Correlation analysis indicates that temperature and vibration-based rolling statistics exhibit stronger relationships with failure events, suggesting gradual degradation patterns prior to breakdowns. Lag features show high correlation with current sensor values, confirming strong temporal dependency in the system.

Overall, the dataset demonstrates physical and statistical consistency, and the engineered features effectively capture temporal behavior, making the data suitable for downstream machine learning modeling.
